## 2. An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following questions. [4 Marks]

```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline
```

### 2.A. Probability that none of the items are defective? [1 Mark]

```python
#stats.binom.pmf(x,n,p)
probForNone = stats.binom.pmf(0,10,0.05)
print(probForNone)
```

### 2.B. Probability that exactly one of the items is defective? [1 Mark]


```python
probForOne = stats.binom.pmf(1,10,0.05)
print(probForOne)
```

### 2.C. Probability that two or fewer of the items are defective? [1 Mark]

```python
probForTwo = stats.binom.pmf(2,10,0.05)
print(probForTwo)
probFor2OrLess = probForNone + probForOne + probForTwo
print("Probability that 2 or less Item will be defective: ",probFor2OrLess)
```

### 2.D. Probability that three or more of the items are defective ? [1 Mark]

```python
prob3OrMore = 1-probFor2OrLess
print(prob3OrMore)
```

### 4 B
#### Null Hypothesis H0: 
 
     Mean of Operating & Closed State is same

#### Alternative Hypothesis Ha:

    Mean of Operating & Closed State is not same and significant difference is available

```python
# 4 C
from statsmodels.stats.weightstats import ztest

s1 = StartupC.loc[StartupC.OperatingState == "Operating", "Funds_in_million"]
s2 = StartupC.loc[StartupC.OperatingState == "Closed", "Funds_in_million"]

# significant level is 5 %
alpha = 0.05

test_stat,p_value = ztest(s1,s2)

if(p_value<alpha):
    print('P Value, ',p_value ,' < ', alpha ,' There is significant difference so we can reject Null Hypothesis H0')
else:
    print('P Value, ',p_value ,' < ', alpha ,' There is not significant difference so we failed to reject Null Hypothesis H0')
```

### 4 G
**Null Hypothesis H0:** The propotion of companies that are operating is same in both categories- Winner & Contestant

**Alternative Hypothesis Ha:** The propotion of companies that are operating is significantly different from each other.

```python
# 4 H
from statsmodels.stats.proportion import proportions_ztest
alpha = 0.05
test_stat , p_val = proportions_ztest([contestant_operat,winner_operat],[contestant,winner])
if(p_val < alpha):
    print('P Value, ',round(p_val,3) ,' < ', alpha ,' There is significant difference so we can reject the Null Hypothesis H0')
else:
    print('P Value, ',round(p_val,3) ,' < ', alpha ,' There is not significant difference so we failed to the reject Null Hypothesis H0')
```
